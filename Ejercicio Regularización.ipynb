# Método del gradiente descendiente (se utilizarán las ecuaciónes normales del modelo de Ridge)

```python
def GradienteDescendiente(X,Y,w0,w,la,eta=0.01):
    for i in range(len(X)):
        error=Y[i]-Y_MPL(w0,w,X[i])
        error=np.copy(error)
        ## Ecuaciones Normales del Modelo de Ridge para Regularización
        w0=w0+eta*(error)
        w[0:]=w[0:]+eta*(error*X[i])-la*w[0:]
        w0=np.asarray(w0)
        w[0:]=np.copy(w[0:])
    return [w0, w]
```
